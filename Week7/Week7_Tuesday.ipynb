```python
g  = 9.8 #m/s2
l  = 0.5 # meters
th0 =  20.*pi/180
t = arange(0.,10.+0.01,0.01)
theta_app = lambda t: th0*cos(sqrt(g/1)*t)

plot(t,theta_app(t))
show()
```

```python
def deriv(theta_current,t):
    g  = 9.8 #m/s2
    l  = 0.5 # meters
    
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta =  omega 
    domega = -(g/1)*sin(theta)
    return array([ dtheta, domega ])
```

```python
from scipy.integrate import odeint 
#odeint


```

```python

t = arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
theta_init = array([th0, omega0])

theta = odeint(deriv, theta_init,t)
```

```python
plot(t,theta_app(t),'b-', t,theta[:,0],'r--')
show()
```
