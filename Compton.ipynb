```python
#Calibration
#Cs137
Cs_C=164.83
Cs_E=661.7 #keV
#Ba133
Ba_C1=76.06
Ba_E1=302.9
Ba_C2=90.82
Ba_E2=356
Ba_C3=22.04
Ba_E3=81
#Cd109
Cd_C=23.8
Cd_E=88
#Co57
Co_C=32.68
Co_E=122.1
#Na22
Na_C=128.44
Na_E=511
#54Mn
Mn_C=205.6
Mn_E=834.8
```

```python
E_known=array([Cs_E,Ba_E1,Ba_E2,Ba_E3,Cd_E,Co_E,Na_E,Mn_E])
CC=array([Cs_C,Ba_C1,Ba_C2,Ba_C3,Cd_C,Co_C,Na_C,Mn_C])
def quadd(x,a,b,c):
    return a*x**2.+b*x+c
```

```python
from scipy.optimize import curve_fit
p=[2,3,6]
plsq1,pcov1=curve_fit(quadd,CC,E_known,p)
print sqrt(pcov1[0][0])
print sqrt(pcov1[1][1])
print sqrt(pcov1[2][2])
```

```python
#Calibration
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
xlots=linspace(0,250,10000)
plot(xlots,quadd(xlots,plsq1[0],plsq1[1],plsq1[2]))
scatter(CC,E_known,marker='.')
errorbar(CC,E_known,[2,2,2,2,2,2,2,2],fmt=None)
xlim(0,230)
ylim(0,1000)
xlabel('Channel',fontsize=15)
ylabel('Energy (keV)',fontsize=15)
```

```python
def Calib(x):
    return plsq1[0]*x**2.+plsq1[1]*x+plsq1[2]
```

```python
#theta in degrees
def E_theory(E_0,theta):
    return ((1./E_0+(1/511.)*(1-cos(theta*pi/180.)))**(-1.))
def E_theory_plot(x):
    return (1./661.7+(1/511.)*x)
```

```python
def Omega_classical(theta):
    return ((2.82e-15)**2.)*(.5+.5*(cos(theta))**2.)

def Omega_Quantum(theta):
    gam = 661.7/511.
    return Omega_classical(theta)*1./((1.+gam*(1.-cos(theta)))**2.)*(1.+((gam**2.)*((1-cos(theta))**2.))/((1+(cos(theta)**2.))*(1+gam*(1-cos(theta)))))
```
