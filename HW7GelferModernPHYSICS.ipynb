$y(x,t)= Asin(kx)cos(wt)$

```python
x = linspace(0,20,100) # range 
f = 0
def wavefunction(x,t):
    A = 2.5 # known Amplitude 
    k = 1 # known wave number
    w = 10*pi # known angular frequency
    f = A*sin(k*x)*cos(w*t) # our given wave function equation
    return f
plot(x,wavefunction(x,.5),'b')
title("Wave function behavior")
xlabel("x")
ylabel("Wave function")
    
```

For t = 1 second
==

```python
plot(x,wavefunction(x,1),'b')
title("Wave function behavior")
xlabel("x")
ylabel("Wave function")
```

$\frac{dy}{dt}=-Awsin(kx)sin(wt)$

```python
x = linspace(0,20,100) # range 
f = 0
def V(x,t): # creating function for velocity 
    A = 2.5 # known Amplitude 
    k = 1 # known wave number
    w = 10*pi # known angular frequency
    f = -A*w*sin(k*x)*sin(w*t) # our given wave function equation
    return f
plot(x,V(x,.5),'b')
title("Transverse Velocity")
xlabel("x")
ylabel("Velocity")
```

for t = 1 second

```python
plot(x,V(x,1),'b')
title("Transverse Velocity")
xlabel("x")
ylabel("Velocity")
```

Problem 7.61 

$\psi1 = A1(\frac{x}{b}e^{\frac{-x^2}{2}})$
$A1 = (\dfrac{4}{\pi b^2})^{\frac{1}{4}}$

$\psi1^2 = (A1(\frac{x}{b}e^{\frac{-x^2}{2}}))^2$

$\psi2 = A2(1-2\frac{x^2}{b^2})e^{\frac{-x^2}{2b^2}}$
$\psi2^2 = (A2(1-2\frac{x^2}{b^2})e^{\frac{-x^2}{2b^2}})^2$

```python
x = linspace(0,3,100) # range 
f = 0
def wave_1(x):
    A = (4/pi)**(.25) # known Amplitude 
    b = 1 # known wave number
    
    f = A*(x/b)*e**((-x**2.)/2.) # our given wave function equation
    return f
plot(x,wave_1(x),'b')
title("wave function 1 vs x ")
xlabel("x")
ylabel("wave function")
```

```python
x = linspace(-2,3,100) # range 
f = 0
def Squared_wave_1(x):
    A = (4/pi)**(.25) # known Amplitude 
    b = 1 # known wave number
    
    f = (A*(x/b)*e**((-x**2.)/2.))**2 # our given wave function equation
    return f
plot(x,Squared_wave_1(x),'b')
title("Probability 1  ")
xlabel("x")
ylabel("Probability 1")
```

```python
x = linspace(-2,4,100) # range 
f = 0
def wave_2(x):
    A = (4/pi)**(.25) # known Amplitude 
    b = 1 # known wave number
    
    f = A*(1-2*((x**2)/(b**2)))*e**((-x**2)/(2*b**2)) # our given wave function equation
    return f
plot(x,wave_2(x),'b')
title("wave function 2 vs x ")
xlabel("x")
ylabel("wave function 2")
```
