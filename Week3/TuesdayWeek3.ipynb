Using Matplotlib to make pretty graphs

```python
#simple line plot
t= arange(0.0, 1.01, 0.01)
c = cos(4*pi*t)
s = sin(4*pi*t)

```

```python
print pi
```

```python
plot(t,c, color="red",linewidth=2.5,label="$cos(4{\pi}t)$")
plot(t,s, color="blue",linewidth=2.5,label="$sin(4{\pi}t)$")
xlabel("time (s)",fontsize=20)
ylabel("voltage (mV)",fontsize=20)
title("my cool graph",fontsize=38)
legend()
grid(True)
savefig("simple_plot.pdf")
show()

```
