Example 1: Analytic Inverse CDF Transform
=

$f(x)= \frac{\sin(x)}{2}$  0 < x < $\pi$ (greater or equal to)

f(x) = 0 otherwise 

```python
x = arange (0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
plot(x,f(x),label='True distribution', lw=3)
plot(x,F(x),label='Cumulative function',lw=3)
xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

```python
# number of samples
N = 10000
#N random numbers between 0 and 1 
z = numpy.random.random_sample(N);
#inverse CDF

Finv = lambda z: arccos(cos(0)-2*z)
#samples x values 
xsamples = Finv(z);
#how do they compare to the original?
hist(xsamples,50,normed=True,label="Samples distirbution")
plot(x,f(x),label='True distribution', lw=3, color='red') 
xlim(0,pi)
ylim(0,0.7)
legend(loc=0)
```

Exercise 2: Exponential function - numerical 
=

```python
import numpy as np
x = arange(0.,5. +0.001,0.01)
CDF = np.cumsum(f(x))/sum(f(x))
N = 10000
z = numpy.random.random_sample(N)
Finv = x[np.searchsorted(CDF,z)]
plot(x,CDF, label = "CDF")
plot(x,f(x),label='True distribution', lw=3, color='red')
hist(Finv,50,normed=True,label="Samples distirbution") 
legend(loc = 0)
show()
```
