```python
# finding speed of light by measuring distances from laser to detector 

x1= 5.75 #m 




t1= 33.2*10**(-9) # seconds 

x2 =  3.1

t2=  24.0*10**(-9)

x3 = 9.307

t3= 45.6*10**(-9)

x4= 0.917

t4= 15.0*10**(-9)
X = array([x1,x2,x3,x4])
T= array([t1,t2,t3,t4])
scatter(T,X)
xlim(0,10**(-7))

show()






```

```python
def Calib(x,c,d):
    return x*c+d
p=[3*10**8,4]
from scipy.optimize import curve_fit  # import the curve fitting function
plsq, pcov = curve_fit(Calib, T, X, p)

xlots=linspace(0,10**(-7),1000)
plot(xlots,Calib(xlots,plsq[0],plsq[1]))
scatter(T,X)
A_=plsq[0]
print A_
```
