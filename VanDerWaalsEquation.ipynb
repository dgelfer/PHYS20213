```python
V = ([1,2,3,4,5,6,7])
def P(v,t):
    return ((8*t)/(3*v-1))-(3/(v**2))
scatter(V,P(V,273))
```

```python
tran=[0.8,0.9,1.0,1.1,1.2]
for t in tran:
    vran=linspace(0.4,7,1000)
    p = [8*t/(3*v-1)-3/v**2 for v in vran]
    plot(vran,p)
    ylim(-1,2)
    xlim(0,7)
    xlabel('V')
    ylabel('p')
    title("Van Der Waal's Equation")


```
