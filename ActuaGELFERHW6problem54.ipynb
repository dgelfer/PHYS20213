```python
x = linspace(-2.,10.,1500) # setting range for wavelength
A0 = 0.25#initial condition for amplitude
lambDa = 4. # wavelength known

Function = zeros(x.shape) # setting parameters for Fourier function

a = lambda n: (2./(n*pi))**2*(1-cos((pi*n)/2)) # Amplitude function

print a(1) #test 

print str("confirmed")
```

```python
def FOURIER(x,n): # Now we can set our actual Fourier function
    A0 = 0.25
    lambDa = 4.
    Function = zeros(x.shape)

    for z in range(n+1): # for every element in the "n" sequence
            if z == 0:
                Function += A0 # since at n = 0 the function has an asmptote, this if statement allows to the set intial amplitude condition
            if z > 0:
                A_nth = (2./(z*pi))**2*(1-cos((pi*z)/2)) # same function as a but in terms of z 
                Function += A_nth*cos(2*z*pi*x/lambDa)
    return Function
print FOURIER(x,3) # test 
```

```python
plot(x,FOURIER(x,3),'.')
title("Fourier series")
xlabel("x")
ylabel("Fourier(x,n)")
```
