electric potential 

$V(x,y) = \frac{kq}{\sqrt{x^2+(y-d)^2}} - \frac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
#make the x-y grid to cover the plane  -2<x<2 and -2<y<2
dx = 0.06
dy = 0.06
x = arange(-2.,2.,dx)
y = arange(-2.,2.,dy)
X,Y =  meshgrid(x,y)
print X.ndim
print Y
# make the function V(x,y)
k = 1; q = 1; d = 1 # note the power of half has a dot at 2

V = (k*q/(X**2+(Y-d)**2)**(1/2.)) - (k*q/(X**2+(Y+d)**2)**(1/2.))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.ocean_r)
# help(Axes3D.plot_surface())
cb = fig.colorbar(p, shrink=0.8)
```

```python
#Represent 3D data by contours
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.50, -1, -2, -4, -8])
contour(X,Y,V,levels,linewidth=3,cmap=cm.gist_heat)
```
